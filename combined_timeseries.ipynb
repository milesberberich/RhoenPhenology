## Packages

```python
import matplotlib.pyplot as plt
import numpy as np
```

## Get DWD data

```python
%store -r sos_np
%store -r eos_np
%store -r los_np

dwd_sos = sos_np
dwd_eos = eos_np
dwd_los = los_np
```

## Get MODIS data

```python
%store -r sos_series
%store -r eos_series
%store -r los_series

modis_sos = sos_series
modis_eos = eos_series
modis_los = los_series
```

## Plotting

```python
import matplotlib.pyplot as plt
import numpy as np
plt.rcParams['font.family'] = 'sans-serif'
plt.rcParams['font.sans-serif'] = ['Gordita', 'Century Gothic', 'Avenir', 'Montserrat', 'DejaVu Sans']
# 1. Define the time axes based on your descriptions
years_dwd = np.arange(1992, 2026)  # 1992 to 2025
years_modis = np.arange(2000, 2025) # 2000 to 2024

# 2. Set up the figure (Default background is white)
fig, ax = plt.subplots(figsize=(22, 7))

# Adjusted colors for better contrast on a white background
color_sos = "#254324"  # Darker Green for SOS
color_eos = '#a04a15'  # Reddish for EOS
color_los = '#7e7a3b'  # Purplish for LOS

# 3. Plot DWD Data (Solid lines, circle markers 'o')
ax.plot(years_dwd, dwd_sos, marker='o', linestyle='-', color=color_sos,
        linewidth=2, markersize=7, label='DWD: Beginning of unfolding')
ax.plot(years_dwd, dwd_eos, marker='o', linestyle='-', color=color_eos,
        linewidth=2, markersize=7, label='DWD: Autumn leave fall')
ax.plot(years_dwd, dwd_los, marker='o', linestyle='-', color=color_los,
        linewidth=2, markersize=7, label='DWD: LOS')

# 4. Plot MODIS Data (Dashed lines, square markers 's')
ax.plot(years_modis, modis_sos, marker='s', linestyle='--', color=color_sos,
        linewidth=2, markersize=6, alpha=0.8, label='MODIS: SOS')
ax.plot(years_modis, modis_eos, marker='s', linestyle='--', color=color_eos,
        linewidth=2, markersize=6, alpha=0.8, label='MODIS: EOS')
ax.plot(years_modis, modis_los, marker='s', linestyle='--', color=color_los,
        linewidth=2, markersize=6, alpha=0.8, label='MODIS: LOS')

# 5. Formatting the Plot (Light Theme)
ax.set_title('Phenology timeseries based on DWD and MODIS',
             fontsize=18, weight='bold', pad=15, color='black')
ax.set_xlabel('Year', fontsize=14, color='black')
ax.set_ylabel('DOY', fontsize=14, color='black')

# Customize axes and grid for light theme
ax.tick_params(colors='black', labelsize=12)
for spine in ax.spines.values():
    spine.set_color('#333333')

# Light gray grid lines so they don't overpower the data
ax.grid(True, color='#e0e0e0', linestyle='--', linewidth=0.7)

ax.legend(loc='upper center', bbox_to_anchor=(0.5, -0.15), ncol=2)
for text in legend.get_texts():
    text.set_color('black')

plt.tight_layout()
plt.show()
```
